```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
top30_pubs = ns_book7[''].value_counts()[:30] #value_count()       30  
top30_pubs
```

```python
top30_pubs_idx = ns_book7[''].isin(top30_pubs.index) # 30   True,   False
top30_pubs_idx
```

```python
top30_pubs_idx.sum()
```

```python
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''])
ax.set_title('  ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''], s=ns_book8[''])
ax.set_title('  ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''],
           linewidths=0.8, edgecolors='k', alpha=0.3,
           s=ns_book8['']*2, c=ns_book8[''])
ax.set_title('  ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sc = ax.scatter(ns_book8[''], ns_book8[''],
           linewidths=0.8, edgecolors='k', alpha=0.3,
           s=ns_book8['']*2, c=ns_book8[''], cmap='jet')
ax.set_title('  ')
fig.colorbar(sc)
fig.show()
```
